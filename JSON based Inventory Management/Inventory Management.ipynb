***Importing necessary Libraries***

```python
import json
import datetime
import random
```

***Menu and Taking User Input***

```python
fd=open('records.json','r')
r=fd.read()
fd.close()
records = json.loads(r)


print('############################################################')
print('#  Welcome to Elite Techno Inventory Management System !   #')

while True:

  print('#  ------------------------------------------------------  #')
  print('# Please Choose one of the following options to continue:  #')
  print('#                                                          #')
  print('# 1. Add Item or New Product to Inventory                  #')
  print('# 2. Get the total no. of Products in the Inventory        #')
  print('# 3. Purchase products from the Inventory                  #')
  print('# 4. Exit                                                  #')
  print('#----------------------------------------------------------#\n')

  inp = int(input('\t\t'))
  print()




  #------------------------Exit Inventory Menu-----------------------------***
    #####################################################################################

  if inp == 4:
    print('----------------------------------------------------------------------------------------------')
    print(' Thank You for using Elite Techno Inventory Inventory Management System ! \n\n Have a Nice Day :)')
    print('----------------------------------------------------------------------------------------------')
    break







    #------------------------Adding New Item to Inventory-----------------------------***
    #####################################################################################

  elif inp == 1:
    
    no_of_items = int(input(' Great! How many Items would you like to add to the Inventory: \t'))
    for i in range(no_of_items):
      prod_id = input('\n Please enter the Product ID of your item:\t')
      print()

      if prod_id in records.keys():
        print(f"So you want to add \'{records[prod_id]['P-Name']}\'. Nice Choice I must say ;)\n")
        quant = int(input('Enter the quantity of the items to be added: \t'))
        print()
        records[prod_id]['Quantity'] += quant

        print(f"Congratulations! {quant} {records[prod_id]['P-Name']}s successfully added to the inventory! ")
      
      else:
        p_Name = input('Please Enter the product name: \t')
        print()
        quant = int(input('Now Enter the quantity: \t'))
        print()
        price = int(input('Finally, Enter the price of each item: \t'))
        print()
        records[prod_id]={'P-Name':p_Name,'Quantity':quant,'Price':price}

        print(f'Congratulations! {quant} {p_Name}s successfully added to the inventory! \n')


      #Updating the inventory after adding items

      js = json.dumps(records)
      fd = open('records.json','w')
      fd.write(js)
      fd.close()


    



  #------------------------No. of Products in Inventory-----------------------------***
    #####################################################################################

  elif inp == 2:
    print('\n Total no. of Products in Inventory = ', len(records.keys())-1, end="\n\n")






  #------------------------Purchasing Products from Inventory-----------------------------***
    #####################################################################################

  elif inp == 3:
    print('####################################################')
    print("#           We're glad to serve you!!              #")
    print('#--------------------------------------------------#')
    print('#       What product do you want to buy?           #')
    print('#                                                  #')
    print('#     (enter the no. against each product !! )     #')
    print('#                                                  #')
    print('# (separate with space incase of multiple products)#')
    print('#--------------------------------------------------#\n\n')
    print('#--------------------------------------------------#')

    for i in range(1,len(records.keys())):
      print(f"# {i}. {records[str(2021000+i)]['P-Name']}".ljust(51) + '#')
    
    print('#--------------------------------------------------#')
    print('\t\t')

    prod = list(input().split())
    quant = []
    sales = {}

    print()

    for i in range(len(prod)):

      quant.append(int(input(f"How many {records[str(2021000+int(prod[i]))]['P-Name']}s would you like to buy? \t")))
      print()

      if (quant[i] > records[str(2021000+int(prod[i]))]['Quantity'] > 0):

        print(f"Sorry!! We don't have enough stock of records{[str(2021000+int(prod[i]))]['P-Name']} !!\n")
        print(f"Do you want to buy {records[str(2021000+int(prod[i]))]['Quantity']} of them ? \n")
        opt = input('Enter Y/N: \t')

        if opt == 'Y':
          quant[i] = records[str(2021000+int(prod[i]))]['Quantity']

        else:
          quant[i] = 0
          continue

      elif records[str(2021000+int(prod[i]))]['Quantity'] == 0: 

        print("Sorry, this product is currently out of stock !! :(")
        continue
      
      else:
        records[str(2021000+int(prod[i]))]['Quantity'] -= quant[i]
        print()


    #Billing Section
        
    print('----------Generating Bill------------\n')
    now = datetime.datetime.now()
    day_now = now.strftime("%Y-%m-%d")
    trans_id = random.randint(100000,900000)
    total = 0

    print('Date and Time:',now.strftime("%Y-%m-%d %H:%M:%S"))
    print('Transaction ID:',trans_id)
    print()

    j = sales["count"] = 1
    sales[trans_id] = {}
    for i in range(len(prod)):

      sales[trans_id][j] = {}

      prod_id = str(2021000+int(prod[i]))

      sales[trans_id][j]["Product ID"] = prod_id
      print('Product ID:',prod_id, end="\n")

      sales[trans_id][j]["Product"] = records[prod_id]['P-Name']
      print('Product:',records[prod_id]['P-Name'], end="\n")

      sales[trans_id][j]["Quantity"] = quant[i]
      print('Quantity:',quant[i], end="\n")
      
      sales[trans_id][j]["Price"] = records[prod_id]['Price']
      print('Price:',records[prod_id]['Price'], end="\n")

      total += (records[prod_id]['Price']*quant[i])
      j += 1

    sales[trans_id]["Amount"] = total
    print('Grand Total:',total, end="\n\n\n")

    sales["count"] += 1

    #Updating the inventory after adding items
    js=json.dumps(records)
    fd=open('records.json','w')
    fd.write(js)
    fd.close()


```
